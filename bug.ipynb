```python
#@title textual

pla_token = "Flan" # @param {type:"string"}
init_token = "girl" # @param {type:"string"}
step = 50   # @param {type:"number"}
train_dir = "/content/drive/MyDrive/Flan" # @param {type:"string"}
output_dir = "/content/drive/MyDrive/textual" # @param {type:"string"}
learn_type = "style" # @param ["style","object"]
model = "runwayml/stable-diffusion-v1-5" # @param ["runwayml/stable-diffusion-v1-5", "stabilityai/stable-diffusion-2-1"] {allow-input: true}

if model=="runwayml/stable-diffusion-v1-5":
  size=512
else:
  size=768

import os
from google.colab import drive


if not drive._os.path.ismount('/content/drive'):
  drive.mount('/content/drive')

os.makedirs(output_dir,exist_ok=True)


!git clone https://github.com/huggingface/diffusers
%cd diffusers
!pip install . -q
%cd /content/diffusers/examples/textual_inversion
!pip install -r ./requirements.txt -q

%env pretrained_model_name_or_path=$model
%env train_data_dir=$train_dir
%env learnable_property=$learn_type
%env placeholder_token=$pla_token
%env initializer_token=$init_token
%env resolution=$size
%env max_train_steps=$step
%env output_dir=$output_dir


!accelerate launch textual_inversion.py \
  --pretrained_model_name_or_path=$pretrained_model_name_or_path \
  --train_data_dir=$train_data_dir \
  --learnable_property=$learnable_property \
  --placeholder_token=$placeholder_token \
  --initializer_token=$initializer_token \
  --resolution=$resolution \
  --train_batch_size=4 \
  --gradient_accumulation_steps=4 \
  --max_train_steps=$max_train_steps \
  --lr_scheduler="constant" \
  --learning_rate=5.0e-04  \
  --lr_warmup_steps=0 \
  --output_dir=$output_dir
```
